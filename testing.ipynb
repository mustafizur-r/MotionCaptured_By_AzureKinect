```python


# Load the .npz file
npy_file_path = 'body_tracking_data_with_metadata.npz'
npy_data = np.load(npy_file_path)

# Checking the content of the file (array keys)
npy_data_keys = npy_data.files  # Get the keys of the stored arrays
print("Keys in npz file:", npy_data_keys)

# Inspect each array in the file
for key in npy_data_keys:
    array_data = npy_data[key]
    print(f"\nKey: '{key}'")
    print(f"Shape: {array_data.shape}")
    print(f"Data type: {array_data.dtype}")
    
    # Check if the array has more than 0 dimensions
    if array_data.ndim > 0:
        # Print a small sample of the array data (first frame or sample)
        print(f"Sample data from '{key}':\n", array_data[0])  # First frame or first element
    else:
        # For 0-dimensional arrays, just print the value
        print(f"Value: {array_data}")
```
